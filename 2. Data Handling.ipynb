# Converting All Formula to pymatgen composition
Note that we have stoichiometric formula and mixed formula

For mixed formula we have a combination of stoichiometric and mass formula inside. There are different representations as well.

(W18O49)0.3(ZnO)0.7 is represented as 0.3 and 0.7 by mass fraction of W18O49 and ZnO respectively, which are represented stoichiometrically

0.9Cu2Se-0.1(Bi0.88Pb0.06Ca0.06CuSeO)-0.04 wt% C shows (Cu2Se)0.9 and (Bi0.88Pb0.06Ca0.06CuSeO)0.1 by stoichiometry, and then 0.04 wt% of C

SnSe0.95 + 1.25 wt% SnCl3 + 0.0675 wt% ErCl3 shows SnSe0.95, SnCl3 and ErCl3  by stoichiometry but 1.25 and 0.0675 are wt%

```python
from pymatgen.core import Composition
import re

# Test formula for the case
formula_1 = "30wt% (W18O49) + 70wt% (ZnO)"
formula_2 = "(Cu2Se)0.9(Bi0.88Pb0.06Ca0.06CuSeO)0.1 - Cu 0.5wt%"
formula_3 = "(Cu2Se)0.9(Bi0.88CuSeO)0.1 - Cu 0.5wt%"
formula_4 = "(Cu2Se)0.9(Bi0.88CuSeO)0.1 + Cu 0.5wt%"
formula_5 = "(Cu2Se)0.9(CuSeO)0.1 + 5wt% Cu"
formula_6 = "O2 + 0.5wt% Cu"
formula_7 = 'SnSe0.95 + 1.25 wt% SnCl3 + 0.25 wt% ErCl3'



# Function to handle both cases where + or - may or may not exist
def split_chemicals(formula):
    if '+' in formula or '-' in formula:
        # Split based on + or -
        chemicals = re.split(r'\s*[\+\-]\s*', formula)
    else:
        # Return the whole formula as one piece
        chemicals = [formula]
    return chemicals

# Function to split chemicals and process items with wt%
def process_formula_to_composition(formula):
    component_dict = {}

    # Split based on + or - first. Note - is a hyphen sign, not minus
    chemicals = split_chemicals(formula)

    for chem in chemicals:
        # Check if wt% is in the item
        if 'wt%' in chem:
            # Extract wt% value and the formula
            wt_match = re.search(r'(\d*\.?\d+)\s*wt%', chem)
            formula_match = re.search(r'\(([\w\d]+)\)|([A-Z][a-z]?[A-Za-z\d]*)', chem)

            if wt_match and formula_match:
                wt_percent = float(wt_match.group(1))

                # Extract the chemical formula
                formula_str = formula_match.group(0)
                comp = Composition(formula_str)
                molar_mass = comp.weight
                mol_fraction = wt_percent / molar_mass

                for element, amount in comp.as_dict().items():
                    component_dict[element] = component_dict.get(element, 0) + amount * mol_fraction 
        else:
            # take this whole part as 100 g or 100wt% of compound 
            wt_percent =  100

            comp = Composition(chem)
            molar_mass = comp.weight

            mol_fraction = wt_percent / molar_mass

            for element, amount in comp.as_dict().items():
                component_dict[element] = component_dict.get(element, 0) + amount * mol_fraction
    
    
    final_comp = Composition(component_dict)

    return final_comp

# Run the function on the test formula
composition_result = process_formula_to_composition(formula_7)

composition_result  # Output the merged pymatgen Composition object
```
